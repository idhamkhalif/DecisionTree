```python
import csv
from sklearn import tree
from sklearn.utils import shuffle
```

```python
#membuka dataset
dataset = []
with open('/content/Training.csv','r') as csvfile:
  r = csv.reader(csvfile,delimiter=',')
  header = next(r)
  for row in r :
    print(row)
    dataset.append(row)
#print(dataset)
```

```python
#acak urutan datset
ds_acak = shuffle(dataset,random_state=2)
print(ds_acak)
```

```python
#pemishan fitur dan label
X = [] #fitur
Y = [] #label
for row in ds_acak:
  X.append([row[0], row[1], row[2], row[3],row[4],row[5],row[6],row[7]])
  Y.append(row[8])
print(X)
print(Y)
```

```python
#melakukan training
clf = tree.DecisionTreeClassifier() #panggil metode
clf = clf.fit(X,Y) #lakukan training
```

```python
#coba prediksi data baru
# Meminta input pengguna untuk prediksi
pregnancies = float(input("Masukkan lama kehamilan: "))
glucose = float(input("Masukkan nilai gula: "))
blood_press= float(input("Masukkan tekanan darah: "))
SkinThickness= float(input("Masukkan SkinThickness: "))
insulin= float(input("Masukkan insulin: "))
BMI = float(input("Masukkan BMI: "))
DiabetesPedigreeFunction= float(input("Masukkan DiabetesPedigreeFunction: "))
age = float(input("Masukkan umur "))

data_baru = [pregnancies, glucose, blood_press, SkinThickness, insulin, BMI, DiabetesPedigreeFunction, age]
klasifikasi = clf.predict([data_baru])

print('data baru :', data_baru)
print('hasil : ', klasifikasi)
```

```python
from sklearn.tree import export_graphviz
import graphviz
from IPython.display import display

# Export pohon keputusan dalam format DOT
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=['Pregnancie', 'Gulcose', 'Blood Pressure', 'Skin Thickness', 'insulin', 'BMI', 'Diabetes Predigre Function', 'Age'],
                           class_names=clf.classes_,
                           filled=True, rounded=True, special_characters=True)

# Buat visualisasi pohon keputusan dari DOT data
graph = graphviz.Source(dot_data)

# Tampilkan visualisasi langsung di output
display(graph)
```
