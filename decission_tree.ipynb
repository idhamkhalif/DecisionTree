```python
import csv
from sklearn import tree
from sklearn.utils import shuffle
```

```python
#membuka dataset
dataset = []
with open('/content/sample_data/data (1)2.csv','r') as csvfile:
  r = csv.reader(csvfile,delimiter=',')
  for row in r :
    print(row)
    dataset.append(row)
#print(dataset)
```

```python
#acak urutan datset
ds_acak = shuffle(dataset,random_state=2)
print(ds_acak)
```

```python
#pemishan fitur dan label
X = [] #fitur
Y = [] #label
for row in ds_acak:
  X.append([row[0], row[1], row[2]])
  Y.append(row[3])
print(X)
print(Y)
```

```python
#melakukan training
clf = tree.DecisionTreeClassifier() #panggil metode
clf = clf.fit(X,Y) #lakukan training
```

```python
#coba prediksi data baru
# Meminta input pengguna untuk prediksi
tinggi_badan = float(input("Masukkan Tinggi Badan (cm): "))
berat_badan = float(input("Masukkan Berat Badan (kg): "))
ukuran_sepatu = float(input("Masukkan Ukuran Sepatu: "))

data_baru = [tinggi_badan, berat_badan, ukuran_sepatu]
klasifikasi = clf.predict([data_baru])

print('data baru :', data_baru)
print('hasil : ', klasifikasi)
```

```python
from sklearn.tree import export_graphviz
import graphviz
from IPython.display import display

# Export pohon keputusan dalam format DOT
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=['Tinggi Badan', 'Berat Badan', 'Ukuran Sepatu'],
                           class_names=clf.classes_,
                           filled=True, rounded=True, special_characters=True)

# Buat visualisasi pohon keputusan dari DOT data
graph = graphviz.Source(dot_data)

# Tampilkan visualisasi langsung di output
display(graph)
```
